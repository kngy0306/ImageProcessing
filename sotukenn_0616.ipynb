```python
# 円検出のテスト画像
img = cv2.imread('./img/opencv-logo.png', 0)
img = cv2.medianBlur(img, 5)

cimg = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB) # 結果出力用の画像

# HoughCircles (グレースケール, 円検出条件, ...)
circles = cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, 1, 20, param1=50, param2=38, minRadius=0, maxRadius=0)

circles = np.uint16(np.around(circles)) # データを整数型に変換

# マーク描写
for i in circles[0, :]:
    cv2.circle(cimg, (i[0], i[1]), i[2], (0, 255, 0), 10) # 緑色で円をマーク
    cv2.circle(cimg, (i[0], i[1]), 2, (255, 0, 0), 10) # 赤色で円をマーク

plt.imshow(cimg)
plt.show()

print(circles.shape)
print(circles)
```

```python
img = cv2.imread("./img/jb.jpg", 0)
img = cv2.medianBlur(img, 9)

cimg = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB) # 結果出力用の画像

# HoughCircles (グレースケール, 円検出条件, ...)
circles = cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, 1, 20, param1=50, param2=38, minRadius=0, maxRadius=0)

circles = np.uint16(np.around(circles)) # データを整数型に変換

# マーク描写
for i in circles[0, :]:
    cv2.circle(cimg, (i[0], i[1]), i[2], (0, 255, 0), 10) # 緑色で円をマーク
    cv2.circle(cimg, (i[0], i[1]), 2, (255, 0, 0), 10) # 赤色で円をマーク

plt.imshow(cimg)
plt.show()

print(circles.shape)
print(circles)
```

```python
jb_gray = cv2.imread("./img/jb.jpg", 0)
jb_gray_blur5 = cv2.medianBlur(jb_gray, 5)

h, w = jb_gray_blur5.shape
size = h * w # 500 x 703
for i in range(size):
    row = int(i / 703)
    column = int(i % 703)
    if jb_gray_blur5[row, column] < 100:
        jb_gray_blur5[row, column] = 0

plt.imshow(jb_gray_blur5)
plt.show()
```
